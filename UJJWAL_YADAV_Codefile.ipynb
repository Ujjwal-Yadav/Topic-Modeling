```
stoplist = set(my_words.split(' '))

texts_data=[]
i=0
for doc_file in text_corpus:
  texts_data.append([])
  for word_element in doc_file.lower().split():
    if word_element not in stoplist:
      texts_data[i].append(word_element)
  i=i+1

for text_file in texts_data:
    for token_element in text_file:
        frequency_of_words[token_element] = frequency_of_words[token_element] + 1


progg_corpus=[]
i=0
for text_file in texts_data:
  progg_corpus.append([])
  for token_element in text_file:
    if (frequency_of_words[token_element]>1):
      progg_corpus[i].append(token_element)
  i=i+1
```

```

dictionary_formed = corpora.Dictionary(progg_corpus)
corpus_final=[]
for text_file in progg_corpus:
  corpus_final.append(dictionary_formed.doc2bow(text_file))
```

```
TF_IDF_model = models.TfidfModel(corpus_final)
corpus_weighted = TF_IDF_model[corpus_final]
#for doc in corpus_tfidf:
#  print(doc)
```

```
my_model = models.LsiModel(corpus_weighted, id2word=dictionary_formed, num_topics=100)
my_model.print_topics(100)
```

```
my_model2 = models.LdaModel(corpus_weighted, id2word=dictionary_formed, num_topics=100)
my_model2.print_topics(100)
```
